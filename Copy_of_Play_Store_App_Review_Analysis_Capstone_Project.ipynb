Data can contain duplicate values,NaN type and wrong datatype so that we need to clean the data

```python
#checking for duplicate values 
playstore_data_df[playstore_data_df.duplicated(subset=['App'], keep='first' )]
```

The above output retrieved 1181 rows of duplicate data. So we need to drop this duplicate data

```python
#droping the  duplicates from App column.
playstore_data_df.drop_duplicates(subset=['App'],keep='first', inplace=True, ignore_index=True)
```

 Here The Rating seem to have a very high number of missing observations.The current ver, Android ver, Type and content rating also have missing values, but they’re pretty negligible compared to Rating.



Replacing all null values with their appropriate values using **aggregate** **functions** such as **mean**,**median**,**mode**



```python
# Replacing the null values in rating column with the median.

median_val =round(playstore_data_df['Rating'].median(),2)
#median_val
playstore_data_df['Rating'].fillna(median_val,inplace=True)
```

Replacing all null values in 'Current version','Android version','Content Rating'and 'Type'columns with mode


```python
playstore_data_df['Current Ver'].fillna(playstore_data_df['Current Ver'].mode()[0],inplace=True)
playstore_data_df['Android Ver'].fillna(playstore_data_df['Android Ver'].mode()[0],inplace=True)
playstore_data_df['Content Rating'].fillna(playstore_data_df['Content Rating'].mode()[0],inplace = True)
playstore_data_df['Content Rating'].fillna(playstore_data_df['Type'].mode()[0],inplace = True)
```
